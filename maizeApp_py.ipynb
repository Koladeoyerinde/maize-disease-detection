```python
import kivy
from kivy.app import App
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.button import Button
from kivy.uix.image import Image
from kivy.uix.label import Label
from kivy.uix.filechooser import FileChooserIconView
import cv2
import numpy as np
import tensorflow as tf
import joblib

# Load CNN model
cnn_model = tf.keras.models.load_model("/content/drive/MyDrive/Maize disease/Maize diseasecnn_model.h5")

# Load SVM model
svm_model = joblib.load("/content/drive/MyDrive/Maize disease/Maize diseasesvm_model.pkl")

# Function to preprocess image
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (128, 128))  # Resize to match CNN input size
    img = img / 255.0  # Normalize
    return np.expand_dims(img, axis=0)

# Function to predict using CNN
def predict_cnn(image_path):
    processed_img = preprocess_image(image_path)
    prediction = cnn_model.predict(processed_img)
    return np.argmax(prediction)

# Function to predict using SVM
def predict_svm(image_path):
    processed_img = preprocess_image(image_path)
    processed_img = processed_img.reshape(1, -1)  # Flatten for SVM
    prediction = svm_model.predict(processed_img)
    return prediction[0]

class MaizeApp(BoxLayout):
    def __init__(self, **kwargs):
        super().__init__(orientation='vertical', **kwargs)
        self.image = Image(size_hint=(1, 0.6))
        self.add_widget(self.image)

        self.file_chooser = FileChooserIconView(size_hint=(1, 0.3))
        self.file_chooser.bind(on_selection=self.selected)
        self.add_widget(self.file_chooser)

        self.result_label = Label(text="Select an image to detect disease", size_hint=(1, 0.1))
        self.add_widget(self.result_label)

        self.detect_button = Button(text="Detect Disease", size_hint=(1, 0.1))
        self.detect_button.bind(on_press=self.detect_disease)
        self.add_widget(self.detect_button)

    def selected(self, file_chooser, selection):
        if selection:
            self.image.source = selection[0]
            self.selected_image = selection[0]

    def detect_disease(self, instance):
        if hasattr(self, 'selected_image'):
            cnn_result = predict_cnn(self.selected_image)
            svm_result = predict_svm(self.selected_image)

            self.result_label.text = f"CNN Prediction: {cnn_result} | SVM Prediction: {svm_result}"
        else:
            self.result_label.text = "Please select an image first"

class MaizeDiseaseApp(App):
    def build(self):
        return MaizeApp()

if __name__ == "__main__":
    MaizeDiseaseApp().run()
```
